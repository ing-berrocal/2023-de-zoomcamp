```python
parquetFile = '../data/green_tripdata_2019-01.parquet'
```

```python
pfData = pf(parquetFile)
data = pfData.to_pandas()
```

```python
data = pd.read_csv('../data/green_tripdata_2019-01.csv')
```

```python
data.dtypes
```

```python
data['lpep_pickup_date'] = pd.to_datetime(data['lpep_pickup_datetime']).dt.date
data['lpep_dropoff_date'] = pd.to_datetime(data['lpep_dropoff_datetime']).dt.date

#pd.to_datetime(data['lpep_dropoff_datetime'], format='%Y-%m-%d').dt.date
data.head()
```

```python
data[(data['lpep_pickup_datetime'] >= '2019-01-15 00:00:00') & (data['lpep_dropoff_datetime'] <= '2019-01-15 23:59:59')]
```

```python
data.groupby(pd.to_datetime(data['lpep_pickup_datetime']).dt.date)['trip_distance'].max()
```
